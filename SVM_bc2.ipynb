```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("/content/bc2.csv")
df.head()
```

```python
df.corr()
```

```python
df.describe().transpose()
```

```python
df.isnull().any()
```

```python
# To treat missing values
df=df.replace("?",np.nan)
df=df.apply(lambda x: x.fillna(x.median()),axis=0)
```

Now we are converting Bare Nuclie column string type to float.

```python
df["Bare Nuclei"]=df["Bare Nuclei"].astype("float64")
df["Bare Nuclei"]
```

```python
from sklearn.model_selection import train_test_split as tts
x=df.drop(["Class","ID"],axis=1)
y=df[["Class"]]
```

```python
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.20,random_state=5)
x_train.shape,y_train.shape
```

Train model with linear kernel.

```python
from sklearn.svm import SVC
model=SVC(C=.1,kernel="linear",gamma=1)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
print(model.score(x_train,y_train))
print(model.score(x_test,y_test))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print("Confusion Matrix:\n",confusion_matrix(y_pred,y_test))
```

In the next steps we will change the kernel and check for accuracies.

```python
model=SVC(kernel="rbf")
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(model.score(x_train,y_train))
print(model.score(x_test,y_test))
```

```python
model=SVC(kernel="poly")
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(model.score(x_train,y_train))
print(model.score(x_test,y_test))
```

```python
model=SVC(kernel="sigmoid")
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(model.score(x_train,y_train))
print(model.score(x_test,y_test))
```
